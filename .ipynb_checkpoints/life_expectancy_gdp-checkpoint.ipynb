```python
# Importing the packages that will be used in this project
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
# Inspecting number of columns and rows using .shape
df.shape
# This inspection show 96 rows and 4 columns
```

```python
# Summaring the data and checking null values
# Summary info
print(df.info())

# Check for missing values
print(df.isnull().sum())
# There are no missing values in the data set
```

```python
# Exploring the DataFrame
# Finding countries in the data
countries = df.Country.unique()
print(countries)
# We find out that there are 6 countries and the name of those countries are: 'Chile' 'China' 'Germany' 'Mexico' 'United States of America' 'Zimbabwe'
```

```python
# Finding Years in the data
years = df.Year.unique()
print(years)
# Years represented in the data are 2000 to 2015
```

```python
# Clean and Prepare Data
# Rename columns for ease
df.rename({'Life expectancy at birth (years)': 'Life_Expectancy'}, axis=1, inplace=True)
print(df.info())
```

```python
# Exploratory Data Analysis
# Overall trend of life expectancy by year
# This line plot shows how life expectancy has changed from 2000 to 2015.
# It helps us compare health progress across countries over time.
sns.lineplot(data=df, x='Year', y='Life_Expectancy', hue='Country')
plt.title('Life Expectancy Over Time')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.show()
```

```python
# GDP trends over time for each country
# This line plot illustrates economic growth by showing changes in GDP over the years.
# It reveals which countries grew fastest and which lagged behind.
sns.lineplot(data=df, x='Year', y='GDP', hue='Country')
plt.title('GDP Over Time')
plt.xlabel('Year')
plt.ylabel('GDP (US Dollars)')
plt.show()
```

```python
# Correlation between GDP and Life Expectancy
# This scatter plot shows the relationship between economic strength and health.
# A positive correlation would suggest that wealthier countries tend to live longer.
sns.scatterplot(data=df, x='GDP', y='Life_Expectancy', hue='Country')
plt.title('GDP vs Life Expectancy')
plt.xlabel('GDP (US Dollars)')
plt.ylabel('Life Expectancy (Years)')
plt.show()
```

```python
# Statistical Insight
# Calculate the correlation between GDP and Life Expectancy
# This shows how strongly these two variables are related.
# A value near 1 means a strong positive relationship, near 0 means no relationship.
correlation = df[['GDP', 'Life_Expectancy']].corr()
print("Correlation matrix:\n", correlation)
```

```python
# Calculate average GDP and life expectancy for each country
# This helps us compare the overall economic and health status of the countries in the dataset.
grouped = df.groupby('Country')[['GDP', 'Life_Expectancy']].mean().sort_values(by='GDP', ascending=False)
print("Country-level averages (sorted by GDP):\n", grouped)
```

```python
# Create a regression plot showing the trend between GDP and Life Expectancy
# This adds a line of best fit, helping to visualize how well GDP predicts life expectancy.
sns.lmplot(data=df, x='GDP', y='Life_Expectancy', hue='Country', aspect=1.5)
plt.title('Linear Regression: GDP vs Life Expectancy')
plt.xlabel('GDP (US Dollars)')
plt.ylabel('Life Expectancy (Years)')
plt.show()
```
